#### 44. Consider a random 10x2 matrix representing cartesian coordinates, convert them to polar coordinates (★★☆)

````python
cc = np.random.randint(-10, 10, (10, 2))
print(cc)

r = np.sqrt(np.sum(cc ** 2, axis = 1, keepdims=True))
theta = np.arctan2(cc[:, 0], cc[:, 1]).reshape(-1,1)

res = np.column_stack([r, theta])
print(res)
````
